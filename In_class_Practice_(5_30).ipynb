```python
import nltk
from nltk.corpus import PlaintextCorpusReader

corpus_root = '/content/drive/MyDrive/Colab Notebooks/Final Project'
wordlists = PlaintextCorpusReader(corpus_root, '.*')

books = wordlists.fileids()
print(books)
```

```python
import nltk
testText = wordlists.words('Description essay_final draft.txt')
single = nltk.text.Text(testText)

findlist = ["I","my","he","his","it","its","we","our"]
single.dispersion_plot(findlist)
```

```python
books = wordlists.fileids()

cfd = nltk.ConditionalFreqDist((fileid, word)
    for fileid in books
    for word in wordlists.words(fileid))

findList = ["can","should","might","may","will","would","could"]
cfd.tabulate(conditions=books, samples=findList)
```

```python
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:7])
    for fileid in wordlists.fileids()
    for w in wordlists.words(fileid)
    for target in ['I','my','he','his']
    if w.lower() == target)
cfd.plot()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Colab Notebooks/Final Project/Description essay_final draft.txt'
file_text = open(file_path, 'r').read()

stop = list(STOPWORDS)

wordcloud = WordCloud(
    background_color="purple",
    width=1500,
    height=1000,
    margin=2,
    stopwords=stop
).generate(file_text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
